```python
import numpy as np
from tqdm import tqdm

import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

## データ読み込み

```python
mnist = datasets.fetch_mldata('MNIST original', data_home='data/src/download/')
```

```python
X, y = mnist.data, mnist.target
X = X / 255.0 #0~1の値に
X = np.array(X/255.0, dtype=np.float32)
y = np.array(y,dtype=np.uint8)

# 10 class分類
class_num = 10
```

```python
#Testを分ける
train_X, test_x, train_Y, test_y = train_test_split(X, y, test_size=0.2, random_state=2)
#TrainとValに分ける
train_x, val_x, train_y, val_y = train_test_split(train_X, train_Y, test_size=0.1, random_state=42)
```

## データセット

```python
row = 8
col = 8
fig = plt.figure(figsize=(row, col))
for i in range(row*col):
    ax = fig.add_subplot(row, col, i+1)
    ax.imshow(train_x[i].reshape(28,28), cmap = cm.gray_r)
    ax.axis('off')
```

```python
print('train data:',train_x.shape,'train label:',train_y.shape)
print('val data:',val_x.shape,'val label:',val_y.shape)
```

## KNN

画素値をそのまま特徴量として扱う.

距離が近いものをk個取ってきて多数決する. 例としてcos距離で実装する.

$$ cos(A, B) = \frac{AB}{|A||B|} $$

```python
def find_k(train_x, train_y, val_x, val_y):
    train_x = train_x/np.linalg.norm(train_x, axis=1, keepdims=True)
    val_x = val_x/np.linalg.norm(val_x, axis=1, keepdims=True)
    
    pred_list = []
    k_list = np.arange(1,4)
    for k in tqdm(k_list):
        dist = np.dot(val_x, train_x.T)
        nearest_k = np.argsort(dist, axis=1)[:,::-1][:,:k]
        labels = train_y[nearest_k]
        label_num = np.sum(np.eye(class_num)[labels], axis=1)
        pred = np.argmax(label_num, axis=1)
        pred_list.append(f1_score(val_y, pred, average='macro'))
        
    top_ind = np.argmax(pred_list)
    best_k = k_list[top_ind]
    print('best k : {}, val score : {}'.format(best_k, pred_list[top_ind]))
    return best_k
```

```python
class KNNEvaluator:
    def __init__(self, test_x, test_y):
        self.test_x = test_x
        self.test_y = test_y
        
    def __call__(self, train_x, train_y, k):
        train_x = train_x/np.linalg.norm(train_x, axis=1, keepdims=True)
        test_x = self.test_x/np.linalg.norm(self.test_x, axis=1, keepdims=True)

        dist = np.dot(test_x, train_x.T)
        nearest_k = np.argsort(dist, axis=1)[:,::-1][:,:k]
        labels = train_y[nearest_k]
        label_num = np.sum(np.eye(class_num)[labels], axis=1)
        pred = np.argmax(label_num, axis=1)
        print('test score : {}'.format(f1_score(self.test_y, pred, average='macro')))
        
crit = KNNEvaluator(test_x,test_y)
```

```python
best_k = find_k(train_x, train_y, val_x, val_y)
```

```python
crit(train_x, train_y, best_k)
```
